實作模糊與邊緣檢測

- 透過 Gaussian Filter 實作模糊操作
- 透過 Sobel Filter 實作邊緣檢測

```python
import cv2
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('data/lena.png')
```

## 邊緣檢測

```python
# 轉為灰階圖片
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

比較 Sobel 如果在 uint8 的情況下做會 overflow 的狀況

```python
# 對 x 方向以包含負數的資料格式 (cv2.CV_16S) 進行 Sobel 邊緣檢測

img_sobel_x = cv2.Sobel(img_grey,
                        cv2.CV_16S,
                        dx=1,
                        dy=0,
                        ksize=3)
```

```python
# 對 x 方向依照比例縮放到所有數值都是非負整數

img_sobel_x = cv2.convertScaleAbs(img_sobel_x)
```

```python
# 對 x 方向直接以非負整數的資料格式 (uint8) 進行 Sobel 邊緣檢測
img_sobel_x_uint8 = np.uint8(img_sobel_x)
```

```python
plt.figure(figsize=(14,14))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(img_sobel_x)
plt.subplot(133)
plt.imshow(img_sobel_x_uint8)
```

### 比較 Sobel 如果在 uint8 的情況下做會 overflow 的狀況

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_sobel_x = cv2.Sobel(img_grey, cv2.CV_16S, dx=1, dy=0, ksize=3)
img_sobel_xx =cv2.Sobel(img_grey, cv2.CV_16S, dx=2, dy=0, ksize=3)
```

```python
plt.figure(figsize=(14,14))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(img_sobel_x)
plt.subplot(133)
plt.imshow(img_sobel_xx)
```
